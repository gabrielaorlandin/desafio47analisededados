```python
import pandas as pd

tabela = pd.read_excel("tabela_alunos.xlsx")

display(tabela)
```

```python
#Registro de alunos que entraram em ordem alfabetica e quantidade

alunos = tabela[["Nome","Atividade"]].groupby(["Nome"]).sum()
display(alunos)
print(f"Entraram {len(alunos.index)} alunos")
```

```python

tabela["Hora"] = tabela["Hora"].astype(str)

# Registro de alunos que entraram após as 18:30 (considerando qualquer hora subsequente)
alunos_entraram_apos_1830 = tabela[
    (tabela["Atividade"] == "Entrou") &
    (tabela["Hora"] > "18:30")
]

# Registro de alunos que saíram antes das 20:30 (considerando qualquer hora anterior)
alunos_sairam_antes_2030 = tabela[
    (tabela["Atividade"] == "Saiu") &
    (tabela["Hora"] < "20:30")
]


print("Alunos que entraram após as 18:30 (considerando qualquer hora subsequente):")
print(alunos_entraram_apos_1830[["Nome", "Hora"]])

print("\nAlunos que saíram antes das 20:30 (considerando qualquer hora anterior):")
print(alunos_sairam_antes_2030[["Nome", "Hora"]])
```
